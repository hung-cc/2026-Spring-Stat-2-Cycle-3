# 青少年行為指標與健康檔案 (Group 14 Summary.pdf)

這份報告是關於第 14 組針對「青少年行為指標與健康檔案（Group 14 Adolescent Behavioral Metrics & Health Profiles）」的研究摘要。

---

## 1. 研究問題與數據背景 (Research Question & Data)
* **核心問題**：目前有吸菸與不吸菸的學生之間，其 BMI 百分位數均值是否存在顯著差異？
* **數據集**：採用 2007 年青少年危險行為調查（YRBS 2007）。
* **研究變數**：
  * **分組變數（自變數）**：目前是否吸菸（Current Cigarette Use），為二元變數（$1 = \text{Yes}$, $0 = \text{No}$）。
  * **反應變數（依變數）**：BMI 百分位數（BMIPCT），為連續變數。

---

## 2. 探索性數據分析亮點 (EDA Highlights)
* **樣本數分佈不均（圖 2-2）**：吸菸習慣的二元分佈顯示兩組樣本數差距顯著。不吸菸組（第 0 組）約有 10,003 人，遠多於吸菸組（第 1 組）的 2,434 人。不過，兩組的樣本量均已足夠龐大，足以進行穩健的獨立樣本 t 檢定。
* **反應變數偏態（圖 2-3）**：整體 BMI 百分位數的原始直方圖呈現明顯的**左偏（Left-Skewed）**分佈，數據大量集中在較高的百分位數區間。

---

## 3. 推論統計與決策 (Statistical Inference)
研究透過「獨立樣本 t 檢定」來推論兩組均值是否相同（參考圖 3-1 吸菸習慣與 BMI 百分位數均值推論決策圖）：
* **數據表現**：
  * 不吸菸組 BMI 均值：**64.39**
  * 吸菸組 BMI 均值：**66.23**
  * 實際均值相差：**1.83**
* **統計檢定值**：
  * t 統計量 (t-stat)：**2.95**
  * P-value：**3.236e-03**
  * 效應量 (Cohen's d)：**0.0667**
* **決策**：**拒絕虛無假設 ($H_0$, $p < 0.05$)**。兩組均值在統計學上存在顯著差異。

---

## 4. 研究結論 (Conclusion)
1. **顯著差異**：統計結果證實，目前有吸菸學生的 BMI 百分位數均值，顯著高於不吸菸的學生（高出約 1.83 個百分位）。
2. **觀察性研究的限制（非因果關係）**：由於本研究屬於觀察性研究，因此這項正相關結果**並不意味著**吸菸會直接導致 BMI 的增加。
3. **潛在干擾變數**：此均值差異可能受到其他底層生活習慣（如飲食習慣、體育活動）或心理變數等干擾因素的影響。

## 3. Statistical Inference
[cite_start]An independent two-sample t-test was conducted to compare the means of the two groups (refer to Figure 3-1)[cite: 13, 15]:
* **Group Metrics**:
  * [cite_start]Non-smokers Mean BMI Percentile: **64.39** [cite: 17, 18]
  * [cite_start]Smokers Mean BMI Percentile: **66.23** [cite: 17, 19]
  * [cite_start]Actual Difference in Means: **1.83** [cite: 46]
* **Test Statistics**:
  * [cite_start]t-statistic (t-stat): **2.95** [cite: 47]
  * [cite_start]P-value: **3.236e-03** [cite: 49]
  * [cite_start]Effect Size (Cohen's d): **0.0667** [cite: 48]
* [cite_start]**Decision**: **Reject $H_0$ ($p < 0.05$)**[cite: 12]. [cite_start]There is a statistically significant difference between the two means[cite: 12].

---

## 4. Conclusion
1. [cite_start]**Significant Difference**: Statistical results indicate that students with current cigarette use have a significantly higher mean BMI percentile (about 1.83 percentiles higher) than non-smokers[cite: 70].
2. [cite_start]**Observational Limitation (No Causation)**: Since this is an observational study, the positive association does not imply that smoking causes an increase in BMI[cite: 71].
3. [cite_start]**Potential Confounders**: This difference may be driven by other underlying lifestyle factors (e.g., dietary habits, physical activity) or psychological variables[cite: 72].
