# Section 1: Variable Coding & Taxonomy / 第 1 節：變數編碼與分類學

To ensure absolute data integrity and structural alignment with the analytical framework, the raw variables extracted from the Youth Risk Behavior Surveillance System (YRBSS) survey have been systematically audited, structured, and operationalized as follows:

為確保絕對的資料完整性，並與分析架構達成結構性對齊，從青少年危險行為監測系統（YRBSS）調查中提取的原始變數已進行系統性的審查、建構與操作化，具體規範如下：

---

## 1. Independent Variable (IV) / 自變項：CurrentCigaretteUse

* **Raw Variable Identification / 原始變數識別**: `CurrentCigaretteUse` (typically mapping to question registry regarding tobacco use frequency in the survey).
    `CurrentCigaretteUse`（通常對應調查中關於過去一個月內吸菸頻率的登記題項）。
* **Data Type / 資料類型**: Numeric / Ordinal (Discrete Categories).
    數值型 / 順序型（離散類別）。
* **Original Coding Schema / 原始編碼規範**:
    The YRBSS records the number of days the respondent smoked cigarettes during the 30 days before the survey using the following multi-categorical scale:
    YRBSS 採用下列多類別量表，紀錄受訪者在調查前 30 天內吸菸的天數：
    * `1`: 0 days / 0 天（從未吸菸）
    * `2`: 1 to 2 days / 1 至 2 天
    * `3`: 3 to 5 days / 3 至 5 天
    * `4`: 6 to 9 days / 6 至 9 天
    * `5`: 10 to 19 days / 10 至 19 天
    * `6`: 20 to 29 days / 20 至 29 天
    * `7`: All 30 days / 全部 30 天
* **Analytical Transformation / 分析轉換 (Recoding)**:
    To directly align with the core research question (*Is the mean BMI percentile different between students who currently use cigarettes and those who do not?*), this ordinal metric is mathematically collapsed into a binary categorical contrast indicator named **`Smoking_Status`**:
    為了直接切合核心研究問題（*吸菸與不吸菸學生之間的平均 BMI 百分位數是否有差異？*），此順序型指標在數學上被簡化並重編碼為名為 **`Smoking_Status`** 的二元類別對照指標：
    * **`0` (Non-Smoker / 不吸菸組)**: Mapped strictly from original code `1` (0 days).
        嚴格對應原始編碼 `1`（0 天）。
    * **`1` (Current Smoker / 吸菸組)**: Aggregated from original codes `2` through `7` ($\ge 1$ day).
        聚合原始編碼 `2` 至 `7`（大於或等於 1 天）。

## 2. Dependent Variable (DV) / 依變項：BMIPCT

* **Raw Variable Identification / 原始變數識別**: `BMIPCT` (Body Mass Index Percentile).
    `BMIPCT`（身體質量指數百分位數）。
* **Data Type / 資料類型**: Numeric / Continuous (Interval-Scale).
    數值型 / 連續型（等距尺度）。
* **Original Coding Schema / 原始編碼規範**:
    Continuous decimal metrics calculated derived from self-reported height and weight, standardizing scores dynamically against CDC national growth reference charts. The value ranges strictly between `0.00` and `100.00`.

  根據受訪者自報的身高與體重計算而得的連續小數指標，並參照美國疾管局（CDC）國家發育標準圖表進行動態標準化。數值嚴格介於 `0.00` 至 `100.00` 之間。
* **Analytical Transformation / 分析轉換**:
    Maintained in its raw continuous interval format. No binning or truncation is applied, ensuring maximum statistical precision and mathematical power during mean destination profiling.

  維持其原始的連續等距格式。不進行任何分箱（Binning）或截斷，以確保在進行均值剖析時能保有最高的統計精確度與數學檢定力。

## 3. Data Integrity & Missing Values / 資料完整性與缺失值規範

During the initialization phase (`01_data-check`), the following taxonomy guidelines are applied before entering the exploratory stage:


在初始化階段（`01_data-check`）中，進入探索性分析前須遵循以下分類學指引：

1.  **System-Missing Codes / 系統缺失值**:
    Any entry recorded as blank or non-numeric (`NaN`, `.` or empty space) in either column is preserved as missing without baseline mutation in Phase 1.

     任一欄位中若紀錄為空白或非數位（如 `NaN`、`.` 或空值），在第一階段皆予以保持原樣，不進行人為底線變更。
3.  **Out-of-Bounds Detection / 邊界外異常檢驗**:
    * Any `CurrentCigaretteUse` values $< 1$ or $> 7$ (excluding `NaN`) are flagged as illegal data anomalies.
        任何 `CurrentCigaretteUse` 數值 $< 1$ 或 $> 7$（不含 `NaN`）皆會被標記為非法資料異常。
    * Any `BMIPCT` values $< 0.00$ or $> 100.00$ (excluding `NaN`) are flagged as operational range outliers.
        任何 `BMIPCT` 數值 $< 0.00$ 或 $> 100.00$（不含 `NaN`）皆會被標記為操作區間異常值。
