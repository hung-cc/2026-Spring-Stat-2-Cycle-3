# Section 5: Statistical Assumptions Considered / 第 5 節：納入考量的統計學假設

Before finalizing inferences, the four classical mathematical assumptions undergirding the Independent Samples $t$-test are methodically audited and defended:
在確立最終統計推論之前，本專案對支持獨立樣本 $t$ 檢定的四個經典數學假設進行了系統性的審查與防禦：

---

## 1. Independence of Observations / 觀測值的獨立性
* **Status / 檢驗狀態**: **🎉 FULLY MET / 完全滿足**
* **Defense / 方法學防禦**: 
  The YRBSS utilizes a stratified, multi-stage cluster sampling design. Each student respondent represents an independent random draw. There are no paired or longitudinal crossovers between the comparison groups.
  YRBSS 採用分層多階段群集抽樣設計，每位受訪高中生皆為獨立隨機抽選。組別之間不存在配對或縱向追蹤的交叉重疊。

---

## 2. Measurement Scale / 測量尺度
* **Status / 檢驗狀態**: **🎉 FULLY MET / 完全滿足**
* **Defense / 方法學防禦**: 
  The dependent variable **`BMIPCT`** is a continuous percentile value scaling strictly between `0.00` and `100.00`, perfectly fulfilling the prerequisite for parametric mean testing.
  依變項 **`BMIPCT`** 為嚴格限制在 `0.00` 至 `100.00` 之間的連續型百分位數，完全滿足參數均值檢定的數學前提。

---

## 3. Normality of Distribution / 常態分佈假設
* **Status / 檢驗狀態**: **⚠️ SLIGHTLY VIOLATED (MITIGATED) / 輕微違反（已完美緩解）**
* **Defense / 方法學防禦**: 
  * **Diagnosis (診斷)**: Visual checks in `02_eda` show a slight right-skew in `BMIPCT` due to population-level overweight trends.
    `02_eda` 的圖表顯示，由於現代青少年整體的過重趨勢，`BMIPCT` 分佈呈現輕微的右偏。
  * **Mitigation (緩解依據)**: According to the **Central Limit Theorem (CLT)**, when the sample size is large ($n \gg 30$ per group), the sampling distribution of the mean approaches normality. With thousands of students, the $t$-test is mathematically robust against this mild skewness.
    根據**中央極限定理 (CLT)**，當每組樣本數極大（$n \gg 30$）時，平均數的抽樣分佈會自動逼近常態分佈。由於本專案樣本高達數千人，$t$ 檢定在數學上對此輕微偏態具備極強的強韌性。

---

## 4. Homogeneity of Variance / 變異數同質性
* **Status / 檢驗狀態**: **⚠️ EMPIRICALLY VIOLATED (BYPASSED) / 經驗性違反（已安全繞過）**
* **Defense / 方法學防禦**: 
  * **Diagnosis (診斷)**: `02_eda` statistics indicate that sample sizes ($n$) and standard deviations ($SD$) between the two groups are unequal (unbalanced design).
    描述性統計顯示，兩組之間的樣本數（$n$）與標準差（$SD$）並不相等（不對稱設計）。
  * **Mitigation (緩解依據)**: The inferential engine defaults to **Welch's $t$-test** instead of Student's $t$-test. By dynamically adjusting the degrees of freedom via the Welch–Satterthwaite equation, unequal variances are handled natively without inflating Type I error ($\alpha$).
    推論流程預設採用 **Welch's $t$ 檢定** 而非傳統 $t$ 檢定。透過動態修正自由度，變異數不相等的現實被安全地原生繞過，不會導致型一錯誤（假陽性）膨脹。
