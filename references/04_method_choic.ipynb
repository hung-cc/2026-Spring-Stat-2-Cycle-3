# Section 4: Statistical Method Choice / 第 4 節：統計分析方法選擇

---

## 1. Selected Framework / 選定架構
The research questions are evaluated using a **Two-Tailed Independent Samples Welch's $t$-test**, complemented by **Cohen's $d$** for effect size measurement.
本研究採用**雙尾獨立樣本 Welch's $t$ 檢定**來評估研究問題，並搭配 **Cohen's $d$** 衡量實質效應量。

---

## 2. Methodological Justification / 方法學依據

The selection of an independent $t$-test is dictated by the variables' properties:
選用獨立樣本 $t$ 檢定是由變數型態決定的硬性規範：
1. **Independent Variable (IV / 自變項)**: `Smoking_Status` is a **Dichotomous Categorical Variable** (Binary: `0` vs `1`), creating two distinct independent groups.
   組別指標為**二元類別變數**（不吸菸組 vs 吸菸組），將樣本切為獨立兩組。
2. **Dependent Variable (DV / 依變項)**: `BMIPCT` is a **Continuous Interval Variable** (`0.00` to `100.00`), which allows for legitimate mean calculations.
   反應指標為從 `0.00` 至 `100.00` 的**連續型等距變數**，符合計算平均值的數學前提。
3. **Objective (分析目標)**: Comparing the means of a continuous variable across exactly two independent groups uniquely maps to the independent $t$-test.
   比較「恰好兩組」獨立群體之間的連續型變數平均值，在統計決策樹中唯一對應的就是獨立樣本 $t$ 檢定。

---

## 3. Why Welch's Variant? / 為什麼選用 Welch 變體？

Traditional Student's $t$-test requires equal variances (Homoscedasticity). This project defaults to **Welch's $t$-test** due to:
傳統 $t$ 檢定要求兩組變異數相等。本專案預設選用 **Welch's $t$ 檢定** 是基於以下現實：

* **Sample Size Asymmetry (樣本量極不對稱)**: The non-smoker group naturally dwarfs the smoker group in a national representative survey. Traditional $t$-tests become highly unreliable under unequal variances with unbalanced samples.
  在全國調查中，不吸菸組人數遠大於吸菸組。當樣本極度不對稱且變異數不相等時，傳統 $t$ 檢定會失效。
* **Mathematical Protection (數學防禦機制)**: Welch's $t$-test handles unequal variances natively by dynamically adjusting the degrees of freedom ($df$) via the **Welch–Satterthwaite equation**:
  Welch's $t$ 檢定不假設變異數相等。它利用 **Welch–Satterthwaite 方程式**動態修正自由度（$df$）：

$$df \approx \frac{\left(\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}\right)^2}{\frac{\left(\frac{s_1^2}{n_1}\right)^2}{n_1-1} + \frac{\left(\frac{s_2^2}{n_2}\right)^2}{n_2-1}}$$

This robustly prevents the inflation of **Type I Error ($\alpha$)** (false positives), ensuring the $P$-value remains statistically unassailable.
這能強韌地防止**型一錯誤（$\alpha$，假陽性）**膨脹，確保最終產出的 $P$ 值在學術上可靠。

---

## 4. Practical Magnitude: Cohen's $d$ / 實質效應量：Cohen's $d$

With a massive sample size ($n > 10,000$), the standard error becomes extremely small, causing even trivial mean differences to trigger a highly significant $P < 0.01$ (sample-size artifact).
由於本專案樣本數龐大（$n > 10,000$），統計上的標準誤極小，容易導致微小的均值差異也產生極度顯著的 $P$ 值（大樣本偽影）。

To address this, **Cohen's $d$** is calculated to shift the focus from mere statistical significance to **practical/clinical significance**, quantifying how many standard deviations separate the two groups:
為此，本研究計算 **Cohen's $d$**，將評估視角提升至**實務/臨床顯著性**，量化兩組學生之間究竟跨越了多少個標準差：

$$d = \frac{\bar{X}_1 - \bar{X}_2}{s_{\text{pooled}}}$$
