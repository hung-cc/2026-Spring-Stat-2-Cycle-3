# Section 3: Response Variable Definition / 第 3 節：反應變數（依變項）定義

To mathematically evaluate the structural impact of adolescent tobacco exposure, this research utilizes **`BMIPCT` (BMI Percentile)** as the primary continuous response variable (Dependent Variable).
為了在數學上評估青少年吸菸暴露的結構性影響，本研究採用 **`BMIPCT`（BMI 百分位數）** 作為核心連續型反應變數（依變項）。

---

## 1. Scientific Definition & Derivation / 科學定義與數據來源

The value of `BMIPCT` is not a direct survey entry; rather, it is a secondary physiological metric derived dynamically through a multi-stage clinical translation:
`BMIPCT` 的數值並非由受訪者直接填寫，而是透過多階段的臨床轉換動態衍生而來的次級生理指標：

1. **Physical Anthropometrics (身體測量)**: The survey capturing self-reported height (meters) and weight (kilograms).
   調查首先蒐集受訪者自報的身高（公尺）與體重（公斤）。
2. **Formula Computation (公式計算)**: The raw Body Mass Index (BMI) is mathematically computed:
   計算原始身體質量指數（BMI）：
   $$\text{BMI} = \frac{\text{Weight (kg)}}{\text{Height (m)}^2}$$
3. **Reference Growth Calibration (對照發育校正)**: 
   This raw index is plotted against the CDC (Centers for Disease Control and Prevention) National Center for Health Statistics growth reference charts.
   將此原始 BMI 指標參照美國疾病管制與預防中心（CDC）國家健康統計中心的發育標準圖表。

---

## 2. Clinical Context & Methodological Justification / 臨床背景與方法學正當性

In adult populations, raw BMI ($\text{kg/m}^2$) is a sufficient metric for body mass stratification. However, in pediatric and adolescent cohorts, raw BMI is **methodologically flawed** because youth undergo radical biological growth spurts where bone density and muscle-to-fat ratios fluctuate wildly by age and biological sex.
在成年人群體中，原始 BMI（$\text{kg/m}^2$）已足夠用於體重分層。然而，在兒童與青少年群體中，使用原始 BMI 在**方法學上是有缺陷的**。因為青少年正處於劇烈的生理發育期，其骨骼密度與肌脂比例會隨著年齡和生物性別產生劇烈波動。

Therefore, **`BMIPCT` (BMI Percentile) is mandated** because it is strictly **age- and sex-specific**:
因此，**必須採用 `BMIPCT`（BMI 百分位數）**，因為它具備嚴格的**年齡與性別校正特性**：

* **Scale Interpretation / 量表分佈解讀**: 
  A `BMIPCT` value of `85.00` implies that the student's adjusted body mass index is higher than 85% of peers of the exact same age and biological sex across the country.
  數值為 `85.00` 代表該學生的校正後體重指標，高於全國高達 85% 同年齡且同生物性別的同儕。
* **Clinical Thresholds / 臨床判定臨界點**:
  * $\text{BMIPCT} < 5.00$: Underweight / 體重過輕
  * $5.00 \le \text{BMIPCT} < 85.00$: Healthy Weight / 健康體重
  * $85.00 \le \text{BMIPCT} < 95.00$: Overweight / 過重
  * $\text{BMIPCT} \ge 95.00$: Obese / 肥胖

---

## 3. Statistical Properties & Scale Scale / 統計本質與測量尺度

From a mathematical modeling perspective, `BMIPCT` satisfies the strict requirements for continuous inferential statistical testing:
從數學建模的角度來看，`BMIPCT` 滿足了連續型推論統計檢定的嚴格要求：

* **Measurement Scale / 測量尺度**: It is a continuous interval variable constrained strictly within the real-number boundary of `0.00` to `100.00`.
  它是一個連續型等距變數，數值範圍嚴格限制在實數邊界 `0.00` 至 `100.00` 之間。
* **Statistical Power Justification / 檢定力正當性**: 
  By maintaining the variable in its raw, continuous percentile format—rather than artificially degrading it into categorical classifications (e.g., binning into "obese" vs "normal")—the data engineering workflow preserves maximum statistical variance and mathematical power during the final Welch's $t$-test execution in `03_inference`.
  透過維持該變數原始的連續百分位數格式，而非人為將其降級為類別分類（例如切編成「肥胖」與「正常」），資料工程流程得以在 `03_inference` 執行最終的 Welch's $t$ 檢定時，完整保留最大的統計變異數與數學檢定力。
