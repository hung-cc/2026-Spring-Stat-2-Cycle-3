# Section 2: Analytical Group Definition / 第 2 節：分析群體定義

This dichotomous classification is achieved by operationalizing and recoding the original multi-categorical ordinal variable **`CurrentCigaretteUse`** into a binary indicator named **`Smoking_Status`**:
此二元分類是透過將原始的多類別順序型變數 **`CurrentCigaretteUse`**，操作化且重編碼為名為 **`Smoking_Status`** 的二元指標來實現：

---

## 1. Group Taxonomy & Operational Criteria / 組別分類學與操作化準則

### A. The Control Group: Non-Smokers / 對照組：不吸菸組
* **Statistical Code / 統計分析編碼**: `Smoking_Status = 0`
* **Visualization Label / 視覺化圖表標籤**: `不吸菸組 (No)`
* **Operational Definition / 操作化定義**: 
  Students who reported absolutely zero frequency of cigarette utilization (`0 days`) within the 30-day temporal window preceding the survey. 
  在調查前 30 天的時間窗口內，回報吸菸使用頻率絕對為零（`0 天`）的學生。
* **Mathematical Mapping / 數學邏輯映射**: 
  $$\text{Smoking\_Status} = 0 \iff \text{CurrentCigaretteUse} = 1$$

### B. The Exposure Group: Current Smokers / 暴露組：吸菸組
* **Statistical Code / 統計分析編碼**: `Smoking_Status = 1`
* **Visualization Label / 視覺化圖表標籤**: `吸菸組 (Yes)`
* **Operational Definition / 操作化定義**: 
  Students who introduced cigarette consumption on at least one day or more (`1 to 30 days`) during the 30-day temporal window preceding the survey. This aggregates casual users through to daily chronic users.
  在調查前 30 天的時間窗口內，回報吸菸天數達一天或一天以上（`1 至 30 天`）的學生。此組別聚合了偶發性吸菸者至每日重度吸菸者。
* **Mathematical Mapping / 數學邏輯映射**: 
  $$\text{Smoking\_Status} = 1 \iff \text{CurrentCigaretteUse} \in \{2, 3, 4, 5, 6, 7\}$$

---

## 2. Summary Transformation Rule / 變數轉換規則摘要

Below is the definitive crosswalk mapping the raw YRBSS survey responses to the analytical groups used in the final $t$-test:
以下為原始 YRBSS 調查回覆與最終 $t$ 檢定所用分析組別之間的明確對照映射表：

| Raw Value <br>(原始回覆數值) | Survey Response Meaning <br>(問卷回覆實際意義) | Recoded Category <br>(重編碼後類別) | Group Assignment <br>(最終分析組別) |
| :---: | :--- | :---: | :--- |
| **`1`** | 0 days (0 天) | **`0`** | **Control Group (Non-Smokers)** <br>對照組（不吸菸組） |
| **`2`** | 1 or 2 days (1 或 2 天) | **`1`** | **Exposure Group (Current Smokers)** <br>暴露組（吸菸組） |
| **`3`** | 3 to 5 days (3 至 5 天) | **`1`** | **Exposure Group (Current Smokers)** <br>暴露組（吸菸組） |
| **`4`** | 6 to 9 days (6 至 9 天) | **`1`** | **Exposure Group (Current Smokers)** <br>暴露組（吸菸組） |
| **`5`** | 10 to 19 days (10 至 19 天) | **`1`** | **Exposure Group (Current Smokers)** <br>暴露組（吸菸組） |
| **`6`** | 20 to 29 days (20 至 29 天) | **`1`** | **Exposure Group (Current Smokers)** <br>暴露組（吸菸組） |
| **`7`** | All 30 days (全部 30 天) | **`1`** | **Exposure Group (Current Smokers)** <br>暴露組（吸菸組） |
