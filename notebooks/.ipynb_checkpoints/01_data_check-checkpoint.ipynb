```python
import warnings
import logging
import os
import platform 

# 1. 忽略 Python 一般警告訊息 (如 Pandas 的 SettingWithCopyWarning)
warnings.filterwarnings('ignore')

# 2. 忽略特定模組的警告 
logging.getLogger('matplotlib.font_manager').setLevel(logging.ERROR)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 3. 顯示中文的設定
def set_ch_font():
    system = platform.system()
    if system == "Windows": 
        plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
    elif system == "Darwin": 
        plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
    plt.rcParams['axes.unicode_minus'] = False

set_ch_font()
sns.set_theme(style="whitegrid", font=plt.rcParams['font.sans-serif'])

# 4. 設定 Pandas 顯示選項 (確保能完整檢視所有欄位)
pd.set_option('display.max_columns', None)

print("✅ 01_data-check 環境初始化完成")
```

```python
import os
import pandas as pd


if 'df' not in locals():
    possible_paths = [
        '../data/raw/yrbs_2007.csv',
        '../data/raw/yrbs_final.csv',
        '../data/processed/cleaned_cycle3.csv'
    ]
    file_loaded = False
    for path in possible_paths:
        if os.path.exists(path):
            df = pd.read_csv(path)
            print(f"🎉 偵測到環境初始化中斷，已自動為此格重新讀取數據：{path} (共 {len(df)} 筆)")
            file_loaded = True
            break
            
    if not file_loaded:
        print("❌ 智慧防禦失敗：找不到實體資料檔案！請先執行最上方的讀檔儲存格。")

# =========================================================================
# 1. 定義 Question 6 研究問題之核心變數
# =========================================================================
target_cols = ['CurrentCigaretteUse', 'BMIPCT']

try:
    # 🚀 關鍵修改：僅選取欄位建立副本供觀察，絕不進行任何整行刪除 (No Deletion)
    df_subset = df[target_cols].copy()

    print("=" * 60)
    print("📊 【Question 6 吸菸者與不吸菸者的平均BMI百分位數是否有差異？】")
    print("=" * 60)

    # 1. 檢查變數資料類型 (Data Types)
    print("【1. 欄位資料類型 (Data Types)】")
    print(df_subset.dtypes)
    print("-" * 50)

    # 2. 檢查缺失值情況 (Missing Value Status)
    print("【2. 缺失值詳細統計 (Missing Values)】")
    missing_count = df_subset.isnull().sum()
    missing_pct = (missing_count / len(df_subset)) * 100
    
    missing_report = pd.DataFrame({
        '原始總筆數': len(df_subset),
        '缺失筆數 (Counts)': missing_count,
        '缺失比例 (%)': missing_pct
    })
    print(missing_report.round(2))
    print("-" * 50)

    # 3. 檢查原始數據的統計特性與分佈範圍 (Descriptive Properties)
    print("【3. 原始數值特徵摘要 (含現存所有數值)】")
    # 透過 describe 可以一眼看出資料的 Count, Mean, Min, Max 等，用來評估變數性質
    print(df_subset.describe().round(2))
    print("-" * 50)

    # 4. 資料合理性與邊界診斷 (僅回報狀況，不進行任何過濾)
    invalid_smoke = df_subset[~df_subset['CurrentCigaretteUse'].isin(range(1, 8)) & df_subset['CurrentCigaretteUse'].notnull()]
    invalid_bmi = df_subset[~df_subset['BMIPCT'].between(0, 100) & df_subset['BMIPCT'].notnull()]

    print("【4. 資料合理性邊界診斷 (異常值抽檢)】")
    print(f"  - 吸菸習慣 (CurrentCigaretteUse) 範圍外異常值: {len(invalid_smoke)} 筆")
    print(f"  - BMI 百分位數 (BMIPCT) 範圍外異常值          : {len(invalid_bmi)} 筆")
    if len(invalid_smoke) == 0 and len(invalid_bmi) == 0:
        print("  => 🎉 診斷通過：現存非空值的數據，皆落在官方定義的合法數值區間內。")
        
    print("\n💡 提示：本儲存格僅進行數據特徵觀測，未對原始 DataFrame 進行任何刪除或清洗。")

except NameError:
    pass
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================================
# 0. 智慧防禦機制：確保 df 變數存在，防止重開 Kernel 報錯
# =========================================================================
if 'df' not in locals():
    possible_paths = ['../data/raw/yrbs_2007.csv', '../data/raw/yrbs_final.csv', '../data/processed/cleaned_cycle3.csv']
    for path in possible_paths:
        if os.path.exists(path):
            df = pd.read_csv(path)
            print(f"🎉 智慧防禦：已自動補齊繪圖所需之原始數據。")
            break

# =========================================================================
# 1. 計算核心變數的交織缺失性質 (不進行任何刪除)
# =========================================================================
target_cols = ['CurrentCigaretteUse', 'BMIPCT']
df_target = df[target_cols]

# 計算個別缺失率
missing_report = pd.DataFrame({
    '缺失筆數': df_target.isnull().sum(),
    '缺失比例 (%)': (df_target.isnull().sum() / len(df_target)) * 100
})

# 深入診斷：兩變數的交織可用狀態 (Properties of Missing Conditions)
both_valid = ((df_target['CurrentCigaretteUse'].notnull()) & (df_target['BMIPCT'].notnull())).sum()
smoke_only = ((df_target['CurrentCigaretteUse'].notnull()) & (df_target['BMIPCT'].isnull())).sum()
bmi_only = ((df_target['CurrentCigaretteUse'].isnull()) & (df_target['BMIPCT'].notnull())).sum()
both_missing = ((df_target['CurrentCigaretteUse'].isnull()) & (df_target['BMIPCT'].isnull())).sum()

# =========================================================================
# 2. 建立專屬視覺化圖表：橫向排版 (1列2欄) 展開，直擊研究主題
# =========================================================================
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- (圖 A) 核心變數個別缺失比例對照 ---
ax_bar = sns.barplot(x=missing_report.index, y='缺失比例 (%)', data=missing_report, 
                     palette='Reds_r', hue=missing_report.index, legend=False, ax=axes[0])
axes[0].set_title('圖 1-1：核心分析變數之個別缺失比例 (%)', fontsize=13, fontweight='bold', pad=10)
axes[0].set_ylabel('缺失比例 (%)', fontsize=11)
axes[0].set_xlabel('專案關鍵變數', fontsize=11)
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['吸菸習慣\n(CurrentCigaretteUse)', 'BMI 百分位數\n(BMIPCT)'])
axes[0].set_ylim(0, max(missing_report['缺失比例 (%)']) + 5)

# 為長條圖加上精確數值標籤
for p in ax_bar.patches:
    ax_bar.annotate(f'{p.get_height():.2f}%', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 8), 
                    textcoords='offset points', fontweight='bold')


# --- (圖 B) 核心變數交織可用狀態結構 (Properties Profile) ---
# 目的：呈現最真實的原始變數性質，看清雙變數在樣本個體上的分佈狀況
status_labels = [
    f'雙變數皆有效\n({both_valid}人)', 
    f'僅吸菸有效 (BMI缺失)\n({smoke_only}人)', 
    f'僅BMI有效 (吸菸缺失)\n({bmi_only}人)', 
    f'雙變數皆缺失\n({both_missing}人)'
]
status_counts = [both_valid, smoke_only, bmi_only, both_missing]

# 篩選掉計數為 0 的分類避免圖表異常
plot_labels = [l for l, c in zip(status_labels, status_counts) if c > 0]
plot_counts = [c for c in status_counts if c > 0]

axes[1].pie(plot_counts, labels=plot_labels, autopct='%1.1f%%', 
            colors=['#66b3ff', '#ffcc99', '#ff9999', '#d3d3d3'], 
            startangle=30, 
            explode=[0.05] + [0] * (len(plot_counts) - 1),
            labeldistance=1.35,
            pctdistance=0.65)
axes[1].set_title('圖 1-2：核心變數樣本交織可用性結構總覽\n(呈現原始資料性質，未經清洗)', fontsize=13, fontweight='bold', pad=10)

# 優化整體佈局
plt.tight_layout()

#  2-in-1 圖組指派給變數
fig_data_check_profile = plt.gcf()
```

```python
import os
import pandas as pd

# =========================================================================
# 0. 智慧防禦機制：確保 df 變數存在，防止 Kernel 斷線重開導致 NameError
# =========================================================================
if 'df' not in locals():
    possible_paths = ['../data/raw/yrbs_2007.csv', '../data/raw/yrbs_final.csv']
    for path in possible_paths:
        if os.path.exists(path):
            df = pd.read_csv(path)
            print(f"🎉 智慧防禦：已自動補齊原始數據 (共 {len(df)} 筆)。")
            break

# =========================================================================
# 1. 🚀 貫徹原則：僅抽取核心變數，不刪除缺失值 (No DropNa)、不進行重編碼
# =========================================================================
target_cols = ['CurrentCigaretteUse', 'BMIPCT']

try:
    # 僅撈取這兩個變數的原始狀況（包含所有 NaN 與原始編碼 1-7）
    df_target_subset = df[target_cols].copy()

    # 2. 自動建立儲存目錄 (防止路徑不存在而報錯)
    processed_dir = '../data/processed/'
    os.makedirs(processed_dir, exist_ok=True)
    
    # 3. 儲存成專案子集資料檔
    processed_path = os.path.join(processed_dir, 'cleaned_cycle3.csv')
    df_target_subset.to_csv(processed_path, index=False, encoding='utf-8-sig')

    print("=" * 60)
    print(f"🚀 【Project Cycle 3 - 01_data-check 完美封裝】")
    print(f"  - 儲存路徑: {processed_path}")
    print(f"  - 原始筆數: {len(df_target_subset)} 筆 (完全保留缺失值與原始編碼)")
    print("=" * 60)

except NameError:
    print("❌ 錯誤：找不到數據物件，請從 Notebook 第一格開始重新執行。")
```
