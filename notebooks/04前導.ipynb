這份清單包含了 75 個與 BMI 顯著相關的指標。為了讓你在論文或報告中能夠清晰地向讀者或教授呈現，我幫你把這 75 個項目依照「公共衛生與行為科學」的標準學術領域**進行完整的**分類與中英對照翻譯。

這樣的分組可以讓你的研究具有極高的邏輯性，以下是完整的翻譯分類：

---

### 📂 類別一：體重覺知與主動體重控制行為（Weight Perception & Weight Control）

*這組指標代表青少年對自己體型的主觀看法，以及他們為了減重所採取的健康或極端行為。*

* **`PerceptionOfWeight`**：自我體型主觀認知（覺得自己極瘦、標準或極胖）
* **`WeightLoss`**：目前是否正在採取減重行動
* **`ConsumeFewerCaloriesToLoseWeight`**：是否透過減少熱量攝取（少吃）來減重
* **`ExerciseToLoseWeight`**：是否透過運動來減重
* **`DietPillsToLoseWeight`**：是否服用減肥藥來控制體重
* **`FastingToLoseWeight`**：是否採取激進節食/禁食（連續24小時以上不吃）來減重
* **`LaxativesToLoseWeight`**：是否服用瀉藥或催吐來控制體重

---

### 📂 類別二：靜態行為與體育活動模式（Sedentary & Physical Activity Behaviors）

*這組指標代表青少年日常能量消耗的作息模式，也是影響肥胖的主要環境型態因子。*

* **`TelevisionWatching`**：平均上學日每天看電視的時間
* **`ComputerUse`**：打電玩或看螢幕等非學業性質的電腦時間
* **`NoerobicExercise`**：未進行有氧運動的頻率
* **`SportsTeamParticipation`**：是否參與學校或社區的運動代表隊
* **`MinutesInPEPlayingSports`**：體育課中實際運動或活動的時間（分鐘）
* **`PEAttendance`**：每週上體育課的天數
* **`PhysicalActivity5OrMoreDays`**：過去 7 天內身體活動達 60 分鐘以上的天數

---

### 📂 類別三：心理健康危機與極端自我傷害（Mental Health & Suicidal Behaviors）

*這組指標實錘了青少年肥胖與心理健康危機、焦慮在群體上的強烈共病現象。*

* **`SadOrHopeless` / `Sad_Binary**`：過去一年內是否曾連續兩週以上感到極度悲傷或絕望
* **`ConsideredSuicide` / `Suicide_Binary**`：過去一年內是否曾認真考慮過自殺（自殺意念）
* **`MadeASuicidePlan`**：過去一年內是否曾擬定過自殺計畫
* **`AttemptedSuicide`**：過去一年內是否曾嘗試自殺（自殺未遂）
* **`InjuriousSuicide`**：自殺未遂且受傷嚴重到需要醫療處理

---

### 📂 類別四：校園暴力衝突與人身安全焦慮（School Violence & Safety Concerns）

*這組指標反映出高 BMI 學生可能在校園人際與霸凌環境中面臨的身體衝突與安全焦慮。*

* **`SafetyConcernsAtSchool`**：過去一個月內是否因校園安全顧慮而不敢去上學
* **`PhysicalFighting`**：過去一年內參與身體打架/鬥毆的總次數
* **`PhysicalFightingAtSchool`**：過去一年內是否曾在校園內與人打架
* **`WeaponCarrying`**：過去一個月內攜帶武器（刀、棍或槍）的天數
* **`WeaponCarryingAtSchool`**：過去一個月內是否曾攜帶武器到校園內
* **`WereThreatenedOrInjuredWithAWeaponOnSchoolProperty`**：在校園內遭受武器威脅或傷害的經歷
* **`StolenOrDamagedYourProperty`**：在校園內財產被偷竊或惡意破壞的次數
* **`InjuredIFight`**：因打架鬥毆而受傷需要就醫的次數

---

### 📂 類別五：菸酒、大麻等物質濫用行為（Substance Use & Risk Behaviors）

*這代表群體中偏差健康行為的「成群出現（Clustering）」特徵，其中包含你的核心自變項。*

* **`CurrentCigaretteUse`**：**（你的核心自變項）** 過去30天當前吸菸頻率
* **`EverCigaretteUse`**：終身是否曾有過吸菸經驗
* **`SmokedMoreThan10Cigarettes`**：單日重度吸菸指標（是否超過10支）
* **`InitiationSmokingWholeCigarette`**：第一次吸完一整支香菸的年齡
* **`EverSmokingCessation`**：過去一年內是否曾嘗試戒菸
* **`CurrentCigarUse`**：過去30天內當前吸食雪茄的頻率
* **`CurrentSmokelessTobaccoUse`**：過去30天當前使用無煙菸品（如嚼菸）的頻率
* **`CurrentSmokelessTobaccoOnSchoolProperty`**：過去30天是否曾在校園內使用無煙菸品
* **`EverMarijuaUse`**：終身是否曾使用過大麻
* **`CurrentMarijuaUse`**：過去30天內當前使用大麻的頻率
* **`InitiationOfMarijuaUse`**：首次嘗試大麻的年齡
* **`CurrentAlcoholUse`**：過去30天內當前的飲酒頻率
* **`CurrentBingeDrinking5OrMore`**：單次飲酒5杯以上的暴飲（Binge Drinking）天數
* **`DrinkAlcoholOnSchoolProperty`**：過去30天是否曾在校園內飲酒
* **`EverUsedLSD`**：是否曾使用過幻覺劑（如 LSD、迷幻藥）
* **`IllegalInjectedDrugUse`**：是否曾使用針頭注射過非法藥物
* **`IllegalDrugsAtSchool`**：過去一年內是否曾在校園內有人向你提供、販售非法藥物
* **`EverSteriodUse`**：是否曾未經處方使用增肌類固醇
* **`InitiationOfAlcoholUse`**：首次飲酒的年齡

---

### 📂 類別六：不安全性行為與防範意識（Sexual Risk Behaviors & HIV Awareness）

*這組項目在行為科學中，常用於評估青少年總體衝動控制力或健康防護意識。*

* **`EverSexualIntercourse`**：終身是否曾發生過性行為
* **`CurrentSexualActivity`** = 過去30天內是否有性行為
* **`MultipleSexPartners`**：終身性伴侶人數（多重性伴侶指標）
* **`CondomUse`**：最近一次性行為是否有使用保險套
* **`BirthControlPillUse`**：性行為時是否有使用口服避孕藥
* **`AlcoholDrugsAndSex`**：最近一次性行為前是否有飲酒或使用藥物
* **`HIVTesting`**：是否曾接受過愛滋病毒（HIV）檢測
* **`TaughtAboutHIV`**：是否曾在學校接受過愛滋病防範教育
* **`ForcedSexualIntercourse`**：是否曾被違反意願強迫發生性行為（遭受性侵經歷）

---

### 📂 類別七：人口統計、常態健康與日常防護（Demographics & Baseline Health）

*這些屬於研究中的結構性背景或常態防護，多用於多元迴歸模型的基礎控制變項。*

* **`BMI_Group`**：根據 BMI 百分位數切出的醫學肥胖分組
* **`WhatIsYourSex`**：生理性別
* **`HowOldAreYou`**：實際年齡
* **`InWhatGradeAreYou`**：目前就讀的年級
* **`WhatIsYourRace`**：你的種族
* **`AreYouHispanicOrLatino`**：是否具有西班牙或拉丁裔血統
* **`HealthInGeneral`**：自我評估的整體健康狀況
* **`Asthma`**：是否曾被診斷患有氣喘
* **`StillHaveAsthma`**：目前是否仍患有氣喘
* **`Sleep`**：上學日的平均睡眠時數
* **`InjuredWhileExercising`**：運動或體育活動時是否曾受傷
* **`CarrotEating`**：過去7天內食用胡蘿蔔的頻率
* **`OtherVegetableEating`**：過去7天內食用其他蔬菜的頻率
* **`SeatBeltUse`**：搭乘汽車時配戴安全帶的頻率
* **`BicyleHelmetUse`**：騎自行車時配戴安全帽的頻率
* **`SunProtection`**：日常採取防曬措施的頻率
* **`SunscreenUse`**：戶外活動使用防曬油的頻率
* **`RidingWithADrinkingDriver`**：是否曾搭乘酒駕者的車輛
* **`HowObtainedCigarettes`**：未成年取得香菸的管道
